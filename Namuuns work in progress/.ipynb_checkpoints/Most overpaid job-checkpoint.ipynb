```python
import pandas as pd
import matplotlib.pyplot as plt

excel_salary = 'salary_data_states.xlsx'
df = pd.read_excel(excel_salary)
```

```python
display(df)
```

```python
df.groupby('JOB_TITLE_SUBGROUP')
```

```python
df.columns
```

```python
df[['JOB_TITLE_SUBGROUP', 'PAID_WAGE_PER_YEAR', 'PREVAILING_WAGE_PER_YEAR']]
```

```python
#difference of offered wage and prevailing wage
df['SALARY_DIFFERENCE'] = df['PAID_WAGE_PER_YEAR'] - df['PREVAILING_WAGE_PER_YEAR']
```

```python
df.groupby(by=['JOB_TITLE_SUBGROUP','SALARY_DIFFERENCE']).agg(len)
```

```python
average_salary_difference = df.groupby('JOB_TITLE_SUBGROUP')['SALARY_DIFFERENCE'].mean()
# Find the job title with the highest average salary difference
most_overpaid_job_title = average_salary_difference.idxmax()

display(pd.DataFrame({"Most Overpaid Job Title": [most_overpaid_job_title], 
                     "Average Salary Difference": [average_salary_difference.max()]}))
```
