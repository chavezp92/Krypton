```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
excel_salary = '/Users/pedrochavez/Desktop/salary_data_states.xlsx'
df = pd.read_excel(excel_salary)
```

```python
df['ACCEPTANCE_RATE'] = (df['PREVAILING_WAGE_PER_YEAR'] / df['PAID_WAGE_PER_YEAR']) * 100
```

```python
mean_acceptance_rate_by_job_title = df.groupby('JOB_TITLE')['ACCEPTANCE_RATE'].mean().reset_index()
```

```python
top_25_job_titles = mean_acceptance_rate_by_job_title.nlargest(25, 'ACCEPTANCE_RATE')
```

```python
print(top_25_job_titles)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_25_job_titles['JOB_TITLE'], top_25_job_titles['ACCEPTANCE_RATE'])
plt.xlabel('Job Title')
plt.ylabel('Mean Acceptance Rate (%)')
plt.title('Top 25 Mean Acceptance Rates by Job Title')
plt.xticks(rotation=90)
plt.show()
```
