```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
excel_salary = '/Users/pedrochavez/Desktop/salary_data_states.xlsx'
df = pd.read_excel(excel_salary)
```

```python
print(df)
```

```python
print(df.columns)
```

```python
df['PAID_WAGE_SUBMITTED'] = pd.to_numeric(df['PAID_WAGE_SUBMITTED'], errors='coerce')
```

```python
mean_paid_wage_by_state = df.groupby('WORK_STATE')['PAID_WAGE_SUBMITTED'].mean()
```

```python
print(mean_paid_wage_by_state)
```

```python
mean_paid_wage_by_state_sorted = mean_paid_wage_by_state.sort_values(ascending=False)
### Now that we have the mean by state we organize in Ascending (Most to Least)
```

```python
print(mean_paid_wage_by_state_sorted)
### Paid Wage Submitted is the Lower Range of Wage Offered to the Applicant.
### In otherwords, the top five states still had a "higher" low point than other states.
### So even if offered the bare minimum in any of the top five states its a far more compelling offer than in other states. 
```
