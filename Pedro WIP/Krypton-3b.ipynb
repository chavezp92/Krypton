```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
excel_salary = '/Users/pedrochavez/Desktop/salary_data_states.xlsx'
df = pd.read_excel(excel_salary)
```

```python
print(df.columns)
```

```python
average_wage_by_employer_title = df.groupby(['EMPLOYER_NAME', 'JOB_TITLE'])['PAID_WAGE_SUBMITTED'].mean()
average_wage_by_employer_title_sorted = average_wage_by_employer_title.sort_values(ascending=False)
```

```python
print(average_wage_by_employer_title_sorted)
```

```python
top_15 = average_wage_by_employer_title_sorted.head(15)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_15.index.get_level_values('EMPLOYER_NAME'), top_15.values)
plt.xlabel('Employer Name')
plt.ylabel('Mean Paid Wage')
plt.title('Top 15 Mean Paid Wage by Employer Name and Job Title')
plt.xticks(rotation=90)
plt.show()
```
