```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
excel_salary = '/Users/pedrochavez/Desktop/salary_data_states.xlsx'
df = pd.read_excel(excel_salary)
```

```python
print(df['JOB_TITLE_SUBGROUP'].sort_values(ascending=False))
```

```python
mean_paid_wage_by_title = (df.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].mean().sort_values(ascending=False))
print(mean_paid_wage_by_title)
```

```python
mean_paid_wage_by_title.plot(kind='bar', figsize=(12, 6), color='green')
plt.title('Mean Paid Wage by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Mean Paid Wage Per Year')
plt.show()
```
