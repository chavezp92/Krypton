# Top 10 Highest Paying States

* California, Washington State, and Massachusetts all have the highest average salaries
  for the jobs listed in this dataset; all states have high costs of living and prominent
  tech sectors.

* Further analysis will need to be done on a broader range of jobs to determine wether these
  trends are focused specifically on jobs in tech and education or the overall job market.

* Nevada has a lower cost of living compared to the other states on this list,
  so the fact that the average overall wage in Nevada is in the top 10 means that they
  have higher salaries overall relative to their cost of living.

```python
# Import libraries and load the data

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
salary_data = 'salary_data_states.xlsx'
df = pd.read_excel(salary_data)
```

```python
# Display the top 5 rows of the DataFrame with all columns
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
# Display the average wage by state
group = df.groupby('WORK_STATE')[['PREVAILING_WAGE_SUBMITTED']].mean()
group.rename(columns={'PREVAILING_WAGE_SUBMITTED': 'AVERAGE_WAGE'},
             inplace=True)

display(group.head())
```

```python
# Display the top 10 states by average wage
group_sorted = group.sort_values('AVERAGE_WAGE', ascending=False)

display(group_sorted.head(10))
```

```python
# Create a bar chart showing the top 10 highest paying states

# Sort data by top 10 highest paying states
top_10 = group_sorted.head(10)

# Create bar chart
top_10['AVERAGE_WAGE'].plot(kind='bar')
plt.title('Top 10 Highest Paying States')
plt.xlabel('State')
plt.ylabel('Average Wage')
plt.xticks(rotation=45)

plt.show()
```
